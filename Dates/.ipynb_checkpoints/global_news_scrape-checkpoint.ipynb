```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url='https://globalnews.ca/news/6859636/ontario-coronavirus-timeline/'
response = requests.get(url)
```

```python
dates = bs(response.text, 'html.parser')
print(dates.prettify())
```

```python
paragraphs = dates.find_all('p')
```

```python
for p in paragraphs:
    if p.strong:
        print(p.get_text())
```

```python
sdates =[]
snews =[]

for p in paragraphs:
    try:
        p = p.get_text()
        event = str(p).split(':',1)
        date = event[0]
        news = event[1]
        sdates.append(date)
        snews.append(news)
        print(date, news)
        print('________')
    except:
        pass
```

```python
sdates
```

```python
del sdates[0:2]
del snews[0:2]
```

```python
data = {
    'Date' : sdates,
    'News' : snews
}
```

```python
dates_df = pd.DataFrame(data, columns = ['Date','News'])
dates_df
```
