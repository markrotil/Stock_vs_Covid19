```python
class stocks(Base):
    __tablename__ = "combined_table"
    index = Column(Integer, primary_key= True)
    date =Column(Date)
    Open =Column(Float)
    High=Column(Float)
    Low=Column(Float)
    Adj_Close =Column(Float)
    Volume = Column(Integer)
    Ticker = Column(String)
```
