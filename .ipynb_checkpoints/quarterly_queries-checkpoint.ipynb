```python
#declare the database we're working with in this notebook
Base = declarative_base()
```

```python
Base.metadata.create_all(engine)
session = Session(bind=engine)
```

```python
#Define a class based on the table format found in the database
class stocks(Base):
    __tablename__ = "everything"
    index = Column(Integer, primary_key= True)
    date =Column(Date)
    Open =Column(Float)
    High=Column(Float)
    Low=Column(Float)
    Adj_Close =Column(Float)
    Volume =Column(Integer)
    Ticker = Column(String)
```

```python
# Testing the connection to the database
dates = session.query(stocks)
for line in dates:
    print(line.date)
    break
```

```python
shopify = session.query(stocks).filter_by(Ticker='SHOP')
for l in shopify: 
    print(l.Ticker, l.date, l.High)
```

```python
q4 = session.query(stocks.Ticker, stocks.Adj_Close)\
.filter(stocks.date <= q4b).filter(stocks.date > q3b).all()
q4
```

```python
q3averages20 = session.query(stocks.Ticker, func.avg(stocks.Adj_Close))\
.group_by(stocks.Ticker)\
.filter(stocks.date <= q4b).filter(stocks.date > q3b).all()
q3averages20
```

```python
q2averages20 = session.query(stocks.Ticker, func.avg(stocks.Adj_Close), func.max(stocks.Adj_Close),\
                          func.min(stocks.Adj_Close))\
.group_by(stocks.Ticker)\
.filter(stocks.date <= q3b).filter(stocks.date > q2b).all()
q2averages20
```

```python
q3averages19 = session.query(stocks.Ticker, func.avg(stocks.Adj_Close), func.max(stocks.Adj_Close),\
                          func.min(stocks.Adj_Close))\
.group_by(stocks.Ticker)\
.filter(stocks.date <= q4a).filter(stocks.date > q3a).all()
q3averages19
```
